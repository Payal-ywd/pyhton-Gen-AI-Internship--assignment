```python
# Problem 1: Mobile Recharge Validation System

# Function to validate recharge
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount >= 50 and amount in valid_plans:
        return "Recharge Successful"
    else:
        return "Invalid Recharge Amount or Plan"

# While loop for retry option
while True:
    amount = int(input("Enter recharge amount: "))
    
    result = validate_recharge(amount)
    print(result)
    
    if result == "Recharge Successful":
        break
    else:
        print("Please try again...\n")
```

```python
# Problem 2: Inventory Reorder Alert System

# Function to check stock
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

# Sample dictionary
inventory = {
    "Product_A": 10,
    "Product_B": 25,
    "Product_C": 5,
    "Product_D": 18
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

# Function to calculate result
def process_result(marks):
    total = 0
    
    # Loop to calculate total
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Sample marks list
marks_list = [60, 45, 70, 55]

result = process_result(marks_list)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    
    fare = base_fare + (distance * per_km)
    
    if peak == "yes":
        fare += fare * 0.25   # 25% extra
    
    return fare

# While loop for retry
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

# Function to check eligibility
def check_eligibility(attendance):
    present_days = 0
    
    # Loop to count present days
    for day in attendance:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance)) * 100
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Sample attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

status = check_eligibility(attendance_list)
print("Attendance Status:", status)
```

```python
# Problem 6: Password Strength Checker

# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    
    # Loop over string
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Input from user
password = input("Enter password: ")

result = check_password(password)
print(result)
```
